```python
# Import Dependencies
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
covid_df = pd.read_csv('COVID-19 Cases.csv')
covid_df.head()
```

```python
# Clean up the df
new_covid_df = pd.DataFrame(covid_df[['Case_Type','Cases','Date','Country_Region','Province_State','Admin2']])
new_covid_df.head()
```

```python
# Rename the Admin2 column
new_covid_df = new_covid_df.rename(columns={'Admin2':'County'})
new_covid_df.head()
```

```python
new_covid_df['Date'] = pd.to_datetime(new_covid_df['Date'])

new_covid_df
```

```python
new_covid_df['Date'] = new_covid_df["Date"].dt.strftime("%m/%d/%y")

new_covid_df.tail()
```

```python
# Create a df with only NC cases
nc_covid_df = new_covid_df.loc[new_covid_df['Province_State'] == 'North Carolina']
nc_covid_df.sort_values(by=['Date','Cases']).head()
```

```python
# See number of cases in North Carolina by Date
nc_covid_df_byDate = pd.DataFrame(nc_covid_df.groupby('Date')['Cases'].sum())
nc_covid_df_byDate.tail(10).sort_values(by='Date',ascending=False)
```

```python
nc_covid_df = nc_covid_df.fillna('-')
nc_covid_df.tail()
```

```python
# Filter/Locate Wake Co., NC deaths by day 
wake_deaths_df = nc_covid_df.loc[nc_covid_df['County'] == 'Wake'][nc_covid_df['Case_Type'] == 'Deaths']

wake_deaths_df.sort_values(by='Date',ascending=False)
```

```python
# Graph Wake Co., NC deaths by day

plt.plot(wake_deaths_df['Date'], wake_deaths_df['Cases'], 'blue','o')
```
